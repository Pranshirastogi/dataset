```python
px.pie(df, names='JobRole', title='Job Role')
```

```python
px.bar(df,x='JobRole', y='BusinessTravel',facet_col='BusinessTravel',color='JobRole', title='Business Travel by Job Role')
```

```python
px.violin(df, x='JobRole', y='MonthlyIncome', color='JobRole',facet_col='Gender', title='Monthly Income by Job Role')
```

```python
px.histogram(df,x='JobRole',y='JobSatisfaction',color='JobRole',title='Job Satisfaction by Job Role')
```

```python
px.bar(df,x='Age',y='MaritalStatus',facet_col='Gender',color='Attrition',title='Marital Status by Age')
```

```python
px.histogram(df,x='Age',y='JobSatisfaction',color='Gender',title='Job Satisfaction by Age')
```

```python
px.scatter(df,x='Age',y='YearsAtCompany',facet_col='Gender',title='Years at Company by Age')
```

```python
px.histogram(df,x='MonthlyIncome',y='PercentSalaryHike',facet_col='Gender',title='Percent Salary Hike by Monthly Income') 
```

```python
px.treemap(df,path=['JobRole','OverTime','JobSatisfaction'],values='MonthlyIncome',title='Job Satisfaction by Job Role and Overtime')
```

```python
px.treemap(df,path=['EducationField','Department','JobRole'],values='MonthlyRate',title='Montly rate by educational field and department')
```
